Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.

```python
import pandas as pd

job_satisfaction_data = {
    "sleep_hours": [7, 8, 6, 9, 5, 4, 10, 3, 2, 1],
    "job_satisfaction": [8, 9, 7, 10, 6, 5, 9, 4, 3, 2],
}

df = pd.DataFrame(job_satisfaction_data)


correlation = df["sleep_hours"].corr(df["job_satisfaction"], method="spearman")


print(f"Spearman's rank correlation coefficient: {correlation:.3f}")

if correlation > 0.7:
    print("There is a strong positive correlation between sleep hours and job satisfaction.")
elif correlation > 0.3:
    print("There is a moderate positive correlation between sleep hours and job satisfaction.")
elif correlation > 0:
    print("There is a weak positive correlation between sleep hours and job satisfaction.")
elif correlation == 0:
    print("There is no monotonic relationship between sleep hours and job satisfaction.")
elif correlation < 0:
    print("There is a weak negative correlation between sleep hours and job satisfaction.")
elif correlation < -0.3:
    print("There is a moderate negative correlation between sleep hours and job satisfaction.")
else:
    print("There is a strong negative correlation between sleep hours and job satisfaction.")
```

Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
between these two variables and compare the results.

```python
import pandas as pd


exercise_bmi_data = {
    "exercise_hours": [2, 3, 4, 5, 1, 6, 7, 8, 9, 10],
    "BMI": [25, 24, 23, 22, 26, 21, 20, 19, 18, 17],
}

df = pd.DataFrame(exercise_bmi_data)

pearson_correlation = df["exercise_hours"].corr(df["BMI"])


spearman_correlation = df["exercise_hours"].corr(df["BMI"], method="spearman")


print(f"Pearson correlation coefficient: {pearson_correlation:.3f}")
print(f"Spearman's rank correlation coefficient: {spearman_correlation:.3f}")


print("Results Comparison:")
if pearson_correlation == spearman_correlation:
    print("The Pearson correlation coefficient and Spearman's rank correlation coefficient are equal.")
elif abs(pearson_correlation - spearman_correlation) < 0.1:
    print("The Pearson correlation coefficient and Spearman's rank correlation coefficient are approximately equal.")
else:
    print("The Pearson correlation coefficient and Spearman's rank correlation coefficient are different.")
```

Q4. A researcher is interested in examining the relationship between the number of hours individuals
spend watching television per day and their level of physical activity. The researcher collected data on
both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
these two variables.

```python
import pandas as pd


tv_physical_activity_data = {
    "TV_hours": [2, 3, 4, 5, 1, 6, 7, 8, 9, 10],
    "physical_activity_level": [3, 2, 1, 4, 2, 5, 4, 3, 2, 1],
}

df = pd.DataFrame(tv_physical_activity_data)

correlation = df["TV_hours"].corr(df["physical_activity_level"])

print(f"Pearson correlation coefficient: {correlation:.3f}")
```

Q5. A survey was conducted to examine the relationship between age and preference for a particular
brand of soft drink. The survey results are shown below:

```python
survey_data = {
    "Age (Years)": [25, 42, 37, 19, 31, 28],
    "Soft Drink Preference": ["Coke", "Pepsi", "Mountain Dew", "Coke", "Pepsi", "Coke"],
}

import pandas as pd
df = pd.DataFrame(survey_data)

crosstab = pd.crosstab(df["Age (Years)"], df["Soft Drink Preference"])

# Print the crosstab
print(crosstab)
```

Q6. A company is interested in examining the relationship between the number of sales calls made per day

30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.
